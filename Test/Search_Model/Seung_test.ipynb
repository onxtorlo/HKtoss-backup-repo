```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
from konlpy.tag import Okt
import re

# 한글 폰트 설정 (matplotlib에서 한글 표시용)
plt.rcParams['font.family'] = 'DejaVu Sans'
```

```python
# 데이터 로드
df = pd.read_csv('2-1.user_input_second.csv')

# 데이터 확인
print("데이터 형태:", df.shape)
print("\n컬럼:", df.columns.tolist())
print("\nsubject 컬럼 샘플:")
print(df['subject'].head())
```

```python
# 텍스트 전처리 함수
def preprocess_text(text):
    if pd.isna(text):
        return ""
    
    # HTML 태그 제거
    text = re.sub('<.*?>', '', str(text))
    
    # 특수문자 제거 (한글, 영문, 숫자, 공백만 유지)
    text = re.sub('[^가-힣a-zA-Z0-9\s]', ' ', text)
    
    # 연속된 공백을 하나로 변환
    text = re.sub('\s+', ' ', text)
    
    return text.strip()

# subject 컬럼 전처리
df['processed_subject'] = df['subject'].apply(preprocess_text)

# 빈 텍스트 확인
print("전처리 후 빈 텍스트 개수:", (df['processed_subject'] == '').sum())
```

```python
# TF-IDF 벡터화
vectorizer = TfidfVectorizer(
    max_features=1000,  # 최대 특성 수
    ngram_range=(1, 2),  # 1-gram과 2-gram 사용
    min_df=1,  # 최소 문서 빈도
    max_df=0.95,  # 최대 문서 빈도 (너무 자주 나오는 단어 제외)
    stop_words=None
)

# 벡터화 실행
tfidf_matrix = vectorizer.fit_transform(df['processed_subject'])

print("TF-IDF 매트릭스 형태:", tfidf_matrix.shape)
print("특성(단어) 개수:", len(vectorizer.get_feature_names_out()))
```

```python
# 코사인 유사도 계산
cosine_sim_matrix = cosine_similarity(tfidf_matrix)

print("코사인 유사도 매트릭스 형태:", cosine_sim_matrix.shape)
print("대각선 값 (자기 자신과의 유사도):", cosine_sim_matrix.diagonal()[:5])
```

```python
# 유사도 매트릭스 시각화
plt.figure(figsize=(12, 10))

# 히트맵 생성
mask = np.triu(np.ones_like(cosine_sim_matrix, dtype=bool))
sns.heatmap(cosine_sim_matrix, 
            mask=mask,
            annot=False, 
            cmap='coolwarm', 
            center=0,
            square=True,
            fmt='.2f')

plt.title('Subject 간 코사인 유사도 히트맵')
plt.xlabel('프로젝트 인덱스')
plt.ylabel('프로젝트 인덱스')
plt.tight_layout()
plt.show()
```

```python
# 유사도 분포 확인
# 대각선 제외 (자기 자신과의 유사도 제외)
similarity_values = cosine_sim_matrix[np.triu_indices_from(cosine_sim_matrix, k=1)]

plt.figure(figsize=(10, 6))
plt.hist(similarity_values, bins=50, alpha=0.7, edgecolor='black')
plt.title('코사인 유사도 분포')
plt.xlabel('코사인 유사도')
plt.ylabel('빈도')
plt.axvline(similarity_values.mean(), color='red', linestyle='--', 
            label=f'평균: {similarity_values.mean():.3f}')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"유사도 통계:")
print(f"평균: {similarity_values.mean():.3f}")
print(f"표준편차: {similarity_values.std():.3f}")
print(f"최대값: {similarity_values.max():.3f}")
print(f"최소값: {similarity_values.min():.3f}")
```

```python
# 가장 유사한 프로젝트 쌍 찾기
def find_most_similar_pairs(cosine_matrix, df, top_n=10):
    # 상삼각 행렬에서 인덱스와 값 추출 (대각선 제외)
    indices = np.triu_indices_from(cosine_matrix, k=1)
    similarities = cosine_matrix[indices]
    
    # 유사도 순으로 정렬
    sorted_indices = np.argsort(similarities)[::-1]
    
    results = []
    for i in range(min(top_n, len(sorted_indices))):
        idx = sorted_indices[i]
        row_idx = indices[0][idx]
        col_idx = indices[1][idx]
        similarity = similarities[idx]
        
        results.append({
            'index_1': row_idx,
            'index_2': col_idx,
            'similarity': similarity,
            'subject_1': df.iloc[row_idx]['subject'][:100] + '...',
            'subject_2': df.iloc[col_idx]['subject'][:100] + '...'
        })
    
    return results

# 가장 유사한 상위 10개 쌍 출력
most_similar = find_most_similar_pairs(cosine_sim_matrix, df, top_n=10)

print("가장 유사한 프로젝트 쌍들:")
print("="*80)
for i, pair in enumerate(most_similar, 1):
    print(f"{i}. 유사도: {pair['similarity']:.3f}")
    print(f"   프로젝트 {pair['index_1']}: {pair['subject_1']}")
    print(f"   프로젝트 {pair['index_2']}: {pair['subject_2']}")
    print("-"*80)
```

```python
# 특정 프로젝트와 가장 유사한 프로젝트들 찾기
def find_similar_to_project(project_index, cosine_matrix, df, top_n=5):
    similarities = cosine_matrix[project_index]
    
    # 자기 자신 제외하고 정렬
    indices = np.argsort(similarities)[::-1][1:top_n+1]
    
    print(f"프로젝트 {project_index}와 가장 유사한 프로젝트들:")
    print(f"기준 프로젝트: {df.iloc[project_index]['subject'][:100]}...")
    print("="*80)
    
    for i, idx in enumerate(indices, 1):
        print(f"{i}. 유사도: {similarities[idx]:.3f}")
        print(f"   {df.iloc[idx]['subject'][:100]}...")
        print("-"*50)

# 첫 번째 프로젝트와 유사한 프로젝트들 찾기
find_similar_to_project(0, cosine_sim_matrix, df)
```

```python
# 유사도 임계값 기반 클러스터링
def find_clusters_by_threshold(cosine_matrix, threshold=0.3):
    clusters = []
    visited = set()
    
    for i in range(len(cosine_matrix)):
        if i in visited:
            continue
            
        cluster = [i]
        visited.add(i)
        
        for j in range(i+1, len(cosine_matrix)):
            if j not in visited and cosine_matrix[i][j] > threshold:
                cluster.append(j)
                visited.add(j)
        
        if len(cluster) > 1:  # 클러스터에 2개 이상의 항목이 있을 때만 추가
            clusters.append(cluster)
    
    return clusters

# 임계값 0.3으로 클러스터링
clusters = find_clusters_by_threshold(cosine_sim_matrix, threshold=0.3)

print(f"유사도 임계값 0.3 기준으로 찾은 클러스터 개수: {len(clusters)}")
for i, cluster in enumerate(clusters):
    print(f"\n클러스터 {i+1} (크기: {len(cluster)}):")
    for idx in cluster:
        print(f"  - {df.iloc[idx]['subject'][:80]}...")
```
