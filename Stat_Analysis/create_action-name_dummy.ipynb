```python
import json
import pandas as pd
import random
from datetime import datetime, timedelta
```

```python
LOG_DATA_PATH = "data\\user-actions_dummy.json"
NAME_DATA_PATH = "..\\Task_recommend\\use_openai\\data\\id_added_dummy_data.json"  # name만 추출할 데이터 파일 경로
```

```python
# 중첩 구조 평탄화해서 읽기, ∵details.name에 값을 넣어야 함
with open(LOG_DATA_PATH, "r", encoding="utf-8") as f:
    t = [json.loads(line.strip()) for line in f if line.strip()]
temp = pd.json_normalize(t)
temp.head(10)
```

```python
with open(NAME_DATA_PATH, "r", encoding="utf-8") as f:
    data = json.load(f)
    print(type(data))

action_names = []

for item in data:
    for category in item.get("recommendedCategories", []):
        for feature in category.get("features", []):
            for action in feature.get("actions", []):
                name = action.get("name")
                if name:
                    action_names.append(name)

print(action_names)
print(len(action_names))
```

```python
# 다시 details로 묶기
temp["details"] = temp.apply(lambda row: {
    "participants": row["details.participants"],
    "state": row["details.state"],
    "name": row["details.name"],
    "startDate": row["details.startDate"],
    "endDate": row["details.endDate"]
}, axis=1)

# 기존 풀린 컬럼 제거
temp.drop(
    columns=["details.participants", "details.state", "details.name", "details.importance", "details.startDate", "details.endDate"],
    inplace=True
)

temp.to_json("data\\final_user-actions_dummy.json", orient="records", force_ascii=False, indent=2)
```
