````python
client = openai.OpenAI()
MODEL_ID = 'gpt-4o-mini'

def create_dummy(df):
    # Read FORMAT prompt
    with open(FORMAT_PROMPT_PATH, "r", encoding="utf-8", errors="replace") as f:
        FORMAT_PROMPT = f.read()

    results = []

    for idx, row in df.iterrows():
        content = row.to_json(force_ascii=False)
        prompt = f"""
        다음은 프로젝트 정보입니다. 이 정보를 바탕으로 **프로젝트의 세부 작업 목록(dummy task list)**을 생성해 주세요.

        각 작업은 다음과 같은 계층 구조로 표현되어야 합니다:
        1. recommendedCategories (3개 이상)
        2. 각 category 하위의 features (3개 이상)
        3. 각 feature 하위의 actions (3개 이상)
        → 모든 작업 항목은 importance (1~5)를 반드시 포함해야 하며, 형식은 JSON만 출력하세요.

        ※ 출력 포맷은 아래 구조 프롬프트를 정확히 따르며, 설명이나 요약 없이 순수 JSON만 출력하세요.

        {content}
        {FORMAT_PROMPT}

        (출력 포맷은 시스템에 등록된 구조 프롬프트를 따릅니다.)
        """
    
        try:
            response = client.chat.completions.create(
                model=MODEL_ID,
                messages=[
                    {"role": "system", "content": SYSTEM_PROMPT},
                    {"role": "user", "content": prompt}
                ],
                temperature=0.3
            )

            output = response.choices[0].message.content.strip()
            
            # JSON 코드블럭 정리
            clean_output = output
            if "```json" in output:
                json_start = output.find("```json") + 7
                json_end = output.rfind("```")
                if json_end > json_start:
                    clean_output = output[json_start:json_end].strip()
            elif "```" in output:
                parts = output.split("```")
                if len(parts) >= 3:  # ```으로 감싸진 경우
                    clean_output = parts[1].strip()
            
            try:
                parsed_output = json.loads(clean_output)
                print(f"성공: {idx + 1}/{len(df)} - JSON 파싱 완료")
                
                results.append({
                    "index": idx,
                    "project": row["text"] if "text" in row else str(row),
                    "raw_output": output,
                    "parsed_data": parsed_output
                })
                
            except json.JSONDecodeError as json_err:
                print(f"JSON 파싱 실패 (인덱스 {idx}): {json_err}")
                print(f"원본 응답 일부: {clean_output[:200]}...")
                
                results.append({
                    "index": idx,
                    "project": row["text"] if "text" in row else str(row),
                    "raw_output": output,
                    "parsing_error": str(json_err),
                    "failed_content": clean_output[:500]
                })
            
            print(f"** {idx + 1} rows data processed! **")
            
        except Exception as e:
            print(f"API 호출 실패 (인덱스 {idx}): {e}")
            
            results.append({
                "index": idx,
                "project": row["text"] if "text" in row else str(row),
                "api_error": str(e)
            })
            continue
    
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = f"dummy_tasks_{timestamp}.json"
    
    try:
        with open(filename, "w", encoding="utf-8") as f:
            json.dump(results, f, ensure_ascii=False, indent=4)
        print(f"** File saved to \"{filename}\" **")
    except Exception as save_error:
        print(f"파일 저장 실패: {save_error}")
        return results

    return results
````

````python
# parsed data 내용만 추출
def quick_extract(input_file, output_file="extracted_tasks.json"):
    with open(input_file, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    tasks = [item["parsed_data"] for item in data if "parsed_data" in item]
    
    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(tasks, f, ensure_ascii=False, indent=2)
    
    print(f"Completed {len(tasks)} tasks → {output_file}")
    return tasks

quick_extract("C:\\Users\\pje03\\Documents\\pja_MLOps\\Task_recommand\\dummy_tasks_20250611_051354.json")
````
