```python
import json

with open("C:\\Users\\user\\Documents\\pja_MLOps\\Task_recommend\\extracted_tasks.json", 'r', encoding='utf-8') as f:
    json_data = json.load(f)

def normalize_json(data):
    normalized = []
    for item in data:
        # 리스트일 경우 첫 번째 요소 사용
        if isinstance(item, list) and len(item) > 0:
            normalized.append(item[0])
        # 딕셔너리일 경우 그대로 추가
        elif isinstance(item, dict):
            normalized.append(item)

    return normalized

normalized_data = normalize_json(json_data)
normalized_json = json.dumps(normalized_data, indent=4, ensure_ascii=False)

print(normalized_json)

with open('normalized_dummy.json', 'w', encoding='utf-8') as file:
    file.write(normalized_json)
```

```python
# workspacdId 수정
import json

def update_id(input_file, output_file="final_normalized_dummy.json"):
    try:
        with open(input_file, "r", encoding="utf-8") as f:
            data = json.load(f)
    except Exception as e:
        print(f"** Error occured!: {e} **")

    for i, item in enumerate(data):
        if "workspaceId" in item:
            item["workspaceId"] = i

    try:
        with open(output_file, "w", encoding="utf-8") as f:
            json.dump(data, f, ensure_ascii=False, indent=4)
    except Exception as e:
        print(f"** Error occured!: {e} **")

    return data

update_id("C:\\Users\\user\\Documents\\pja_MLOps\\Task_recommend\\normalized_dummy.json")
```
